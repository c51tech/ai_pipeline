```python
import datetime

start_time = datetime.datetime.now()

list_url_pattern = 'http://kin.naver.com/qna/expertAnswerList.nhn?dirId=%d&page=%d'
dir_ids = [
    60801, 60802, 60803, 60804
]

domain = 'http://kin.naver.com'

for dir_id in dir_ids:
    for page in range(99,0,-1):
        print(dir_id, page, datetime.datetime.now())
        
        list_url = list_url_pattern % (dir_id, page)
        html = request.urlopen(list_url).read().decode('utf-8', 'ignore')
        p = re.compile('href=\"(/qna/detail.nhn.+)\" ')
        detail_paths = re.findall(p, html)
        
        detail_urls = []
        for detail_path in reversed(detail_paths):
            detail_urls.append(domain + detail_path)
            
        task = web_tasks.Web2Mongo()

        task.urls = detail_urls
        task.mg_collection = 'test_naver_kin'
        task.max_sleep_sec = 5

        task.run()
        
print(datetime.datetime.now(), 'elapsed_time:', datetime.datetime.now() - start_time)

```
